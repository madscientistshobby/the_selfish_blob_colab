```python
import io
import os
import shutil
import urllib.request
import zipfile
from pathlib import Path

CONTENT = Path("/content")
GAME = "the_selfish_blob"
game_dir = CONTENT / GAME
ZIP_URL = (
    "https://github.com/madscientistshobby/the_selfish_blob/"
    "archive/refs/heads/main.zip"
)


def ensure_game_files():
    if game_dir.is_dir() and (game_dir / "Universe.py").is_file():
        return
    if game_dir.is_dir():
        shutil.rmtree(game_dir)
    print("Downloading game (zip)...")
    req = urllib.request.Request(
        ZIP_URL,
        headers={"User-Agent": "colab-the_selfish_blob/1.0"},
    )
    with urllib.request.urlopen(req, timeout=180) as resp:
        data = resp.read()
    with zipfile.ZipFile(io.BytesIO(data)) as zf:
        zf.extractall(CONTENT)
    extracted = CONTENT / "the_selfish_blob-main"
    if not extracted.is_dir():
        raise RuntimeError(
            "Zip layout unexpected (expected folder the_selfish_blob-main). "
            "Check default branch is main on GitHub."
        )
    shutil.move(str(extracted), str(game_dir))
    print("Game files ready at:", game_dir)


ensure_game_files()

# colab_run.py next to this notebook, or fetch from raw
candidates = [
    CONTENT / "the_selfish_blob_colab" / "colab_run.py",
    Path.cwd() / "colab_run.py",
]
src = next((p for p in candidates if p.is_file()), None)
dst = game_dir / "colab_run.py"
if src is not None:
    shutil.copy(src, dst)
else:
    raw = (
        "https://raw.githubusercontent.com/madscientistshobby/"
        "the_selfish_blob_colab/main/colab_run.py"
    )
    urllib.request.urlretrieve(raw, dst)

os.chdir(game_dir)
print("cwd:", os.getcwd())

```
