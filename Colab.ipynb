```python
import os
import shutil
import subprocess
import urllib.request
from pathlib import Path

CONTENT = Path("/content")
GAME = "the_selfish_blob"
GAME_URL = "https://github.com/madscientistshobby/the_selfish_blob.git"
game_dir = CONTENT / GAME

if not game_dir.is_dir():
    subprocess.run(["git", "clone", GAME_URL], cwd=str(CONTENT), check=True)
else:
    subprocess.run(["git", "-C", str(game_dir), "pull", "-q"], check=False)

# colab_run.py: Colab이 이 저장소를 체크아웃한 경우 옆에 있음
candidates = [
    CONTENT / "the_selfish_blob_colab" / "colab_run.py",
    Path.cwd() / "colab_run.py",
]
src = next((p for p in candidates if p.is_file()), None)
dst = game_dir / "colab_run.py"
if src is not None:
    shutil.copy(src, dst)
else:
    raw = "https://raw.githubusercontent.com/madscientistshobby/the_selfish_blob_colab/main/colab_run.py"
    urllib.request.urlretrieve(raw, dst)

os.chdir(game_dir)
print("cwd:", os.getcwd())
```
